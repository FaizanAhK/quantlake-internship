# Merging DataFrames

```python
customers = pd.DataFrame({
    'CustomerID': [1, 2, 3],
    'Name': ['Faizan', 'Ahmad', 'Khan']
})
```

```python
orders = pd.DataFrame({
    'OrderID': [101, 102, 103],
    'CustomerID': [1, 2, 3],
    'Amount': [300, 400, 500]
})
```

```python
# Merge

pd.merge(customers, orders, on = 'CustomerID', how = 'inner')
```
